```python
import warnings
for warn in [UserWarning, FutureWarning]: warnings.filterwarnings("ignore", category = warn)
import torchvision
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from PIL import Image
```

```python
image_path = "C:/Users/lbask/OneDrive/Документы/home/imgs/dl-hww.jpg"
image = Image.open(image_path)
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.6),
    transforms.RandomAffine(degrees=(0, 180), translate=(0.1, 0.3)),
    transforms.RandomSolarize(threshold=156, p=0.5),
    transforms.ColorJitter(brightness=.5, hue=.3)
])

plt.figure(figsize=(2, 2))
plt.imshow(image)
plt.title("исходное изображение:",fontsize=9)
plt.axis("off")
plt.show()

plt.figure(figsize=(16, 9))
for i in range(10):
    augmented_image = transform(image)
    plt.subplot(1, 11, i + 2)
    plt.imshow(augmented_image)
    plt.axis("off")

plt.show()
```

### Вывод 
К изображению были применены 5 различных аугментаций: Resize, RandomHorizontalFlip,RandomAffine,RandomSolarize и ColorJitter. Это помогло увеличить вариативность данных. В свою очередь увеличение вариативности данных может помочь улучшить обобщающую способность модели, которая будет на них обучаться, и уменьшить переобучение.
